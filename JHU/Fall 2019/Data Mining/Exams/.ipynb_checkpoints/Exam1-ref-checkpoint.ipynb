2. Calculate the sample averages of each column (2 pts)

```python
n = data.shape[0]
```

```python
## By hand
avg = np.sum(data, axis=0) / n
avg
```

```python
## By existing function from numpy (just to compare the result, same below)
np.mean(data, axis=0)
```

3. Calculate the sample variances of each column (2 pts)
<br/><br/>
<font color="red">**NOTE**: Need to do it by hand.</font>

```python
## By hand (divided by n-1)
var = np.sum((data-avg)**2, axis=0) / (n-1)
var
```

```python
## By existing function (divided by n-1)
np.var(data, axis=0, ddof=1)
```

```python
## By existing function (divided by n-1)
np.var(data, axis=0) * n / (n-1)
```

```python
## By existing function (divided by n)
np.var(data, axis=0)
```

4. Calculate the skewness of each column (2 pts)
<br/><br/>
<font color="red">**NOTE**: Need to do it by hand, either way of correction for bias is fine.</font>

```python
## By hand 
m3 = np.sum((data-avg)**3, axis=0) / n
skew = m3 / var**(3/2)
skew
```

```python
## By hand (alternatively)
m2 = np.sum((data-avg)**2, axis=0) / n
skew = m3/m2**(3/2)
skew
```

```python
## By existing function (reference)
from scipy import stats

stats.skew(data, axis=0)
```

```python
## By hand (alternatively)
m2 = np.sum((data-avg)**2, axis=0) / n
skew = (np.sqrt(n*(n-1))/(n-2)) * (m3/m2**(3/2))
skew
```

```python
## By existing function (reference)
from scipy import stats

stats.skew(data, axis=0, bias=False)
```

5. Calculate the covariances (2 pts)
<br/><br/>
<font color="red">**NOTE**: Need to do it by hand.</font>

```python
## By hand (seperately)
cov_xy = np.sum((data[:,0]-avg[0])*(data[:,1]-avg[1]), axis=0) / (n-1)
cov_xz = np.sum((data[:,0]-avg[0])*(data[:,2]-avg[2]), axis=0) / (n-1)
cov_yz = np.sum((data[:,1]-avg[1])*(data[:,2]-avg[2]), axis=0) / (n-1)

print('cov(x,y): ', cov_xy)
print('cov(x,z): ', cov_xz)
print('cov(y,z): ', cov_yz)
```

```python
## By hand (covariance matrix)
(data-avg).T @ (data-avg) / (n-1)
```

```python
## By existing function (divided by n-1)
np.cov(data.T)
```

```python
## By existing function (divided by n)
np.cov(data.T, ddof=0)
```

## Problem 2

Use generalized linear regression to fit $z$ using the following family of functions, 

> $\displaystyle f(x,y;\boldsymbol{\beta}) = \sum_k \beta_k \,\phi_k(x,y)$
> <br>
> where
> <br>
> $\displaystyle \phi_0(x,y) = 1$
> <br>
> $\displaystyle \phi_1(x,y) = x$
> <br>
> $\displaystyle \phi_2(x,y) = y$
> <br>
> $\displaystyle \phi_3(x,y) = \sin(x)$
> <br>
> $\displaystyle \phi_4(x,y) = \cos(x)$
> <br>
> $\displaystyle \phi_5(x,y) = \sin(y)$
> <br>
> $\displaystyle \phi_6(x,y) = \cos(y)$


1. Construct the design matrix `X` from the given dataframe `df` for the above problem (3 pts)
2. Derive the best $\hat{\boldsymbol{\beta}}$ coefficients (3 pts)
3. Derive the estimated values $\hat{z}_i=f(x_i,y_i;\hat{\boldsymbol{\beta}})$ for each data point $i$ (2 pts)
3. Plot the histogram of the $\{(z_i\!-\!\hat{z}_i)\}$ residuals (2 pts)
4. Based on the best fit, estimate the $\hat{z}$ at the origin of the $x\!-\!y$ plane, i.e., calculate the $f(0,0;\hat{\boldsymbol{\beta}})$ value (3 pts)

Total of 13 points

1. Construct the design matrix `X` from the given dataframe `df` for the above problem (3 pts)
<br/><br/>
<font color="red">**NOTE**: Use $x$ and $y$ to fit $z$.</font>

```python
## Design matrix
X = np.ones((n,7))
X[:,1] = df.x
X[:,2] = df.y
X[:,3] = np.sin(df.x)
X[:,4] = np.cos(df.x)
X[:,5] = np.sin(df.y)
X[:,6] = np.cos(df.y)
X.shape
```

2. Derive the best $\hat{\boldsymbol{\beta}}$ coefficients (3 pts)

```python
## Least square estimates
z = np.array(df.z)
betahat = np.linalg.inv(X.T @ X) @ X.T @ z
betahat
```

3. Derive the estimated values $\hat{z}_i=f(x_i,y_i;\hat{\boldsymbol{\beta}})$ for each data point $i$ (2 pts)

```python
## Fitted values
zhat = X @ betahat
zhat.shape
```

4. Plot the histogram of the $\{(z_i\!-\!\hat{z}_i)\}$ residuals (2 pts)

```python
residuals = z - zhat
hist(residuals, bins=50);
xlabel('residuals');
```

5. Based on the best fit, estimate the $\hat{z}$ at the origin of the $x\!-\!y$ plane, i.e., calculate the $f(0,0;\hat{\boldsymbol{\beta}})$ value (3 pts)

```python
## New X
Xnew = np.zeros((1,7))
Xnew[:,0] = 1
Xnew[:,4] = 1
Xnew[:,6] = 1
Xnew
```

```python
## Estimated value for new X
Xnew @ betahat
```

```python
## Alternatively by hand
betahat[0] + betahat[4] * 1 + betahat[6] * 1
```
